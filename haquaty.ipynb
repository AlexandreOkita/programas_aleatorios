```python
import pandas as pd
sales = pd.read_csv('sales_train_bem_pequeno.csv')
sales = sales.drop(['Unnamed: 0', 'date_block_num', 'item_cnt_day'], axis=1)
sales.insert(3, 'item_group', 0)
```

```python
def main(user, dados, catalogo):
    compras = cliente(user, dados[1], dados[2], dados[3])
    shallow = juntar(compras)
    grupo = agrupar(shallow, catalogo, compras)
    porcent = porcentagem(grupo)
    print(porcent)
    pref = maior_freq(porcent)
    carzinho = carrinho(pref, shallow, catalogo)
    print(carzinho)

def cliente(id, user, id_item, id_grupo):
    compras = [[], []]
    for i in range(len(user)):
        if user[i] == id:
            compras[0].append(id_item[i])
            compras[1].append(id_grupo[i])
    return compras

def juntar(item):
    quant = []
    produto = []
    for i in item[0]:
        if i not in produto:
            produto.append(i)
            sum = 0
            for j in item[0]:
                if j == i:
                    sum += 1
            quant.append(sum)
    for i in range(len(produto)):
        for j in range(len(produto)):
            if quant[j] < quant[i]:
                produto[i], produto[j] = produto[j], produto[i]
                quant[i], quant[j] = quant[j], quant[i]
    compras = {}
    for i in range(len(produto)):
        compras[produto[i]] = quant[i]
    return(compras)

def agrupar(quant, grupos, compras):
    freq = {}
    for i in quant:
        i_grupo = compras[0].index(i)
        grupo = compras[1][i_grupo]
        soma = 0
        for j in grupos[grupo]:
            soma += quant[j]
        freq[grupo] = soma
    return freq

def porcentagem(freq):
    total = 0
    for i in freq:
        total += freq[i]
    aux = {}
    for i in freq:
        aux[i] = freq[i]*100/total
    porcent = {}
    for i in aux:
        max = 0
        for j in aux:
            if aux[i] > i and aux[i] not in porcent:
                max = aux[i]
        porcent[i] = max
    return porcent

def maior_freq(porcent):
    itens = []
    aparece = 5*len(porcent)//10
    n = 0
    for i in porcent:
        itens.append(i)
        n += 1
        if n == aparece:
            break
    return itens
    
def carrinho(itens, quant, grupos):
    carrinho = {}
    for i in itens:
        for j in quant:
            if j in grupos[i]:
                carrinho[i] = j
                break
    return carrinho

def alternativas(grupo, carrinho, grupos):
    alt = []
    for i in grupos[grupo]:
        if i != carrinho[grupo]:
            alt.append(i)
    return(alt)

main(27, array, produtos)
```
