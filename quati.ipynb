```python
import pandas as pd
sales = pd.read_csv('sales_train_bem_pequeno.csv')
sales = sales.drop(['Unnamed: 0', 'date_block_num', 'item_cnt_day'], axis=1)
sales.insert(3, 'item_group', 0)
sales.head()
```

```python
def define_grupo(item_id, grupo, ds):
    ds.loc[ds.item_id == item_id, 'item_group'] = grupo
```

```python
define_grupo(2552, 1, sales)
define_grupo(2555, 2, sales)
define_grupo(3186, 0, sales)
define_grupo(11041, 0, sales)
define_grupo(6495, 3, sales)
define_grupo(7130, 3, sales)
define_grupo(22014, 4, sales)
sales.head()
```
